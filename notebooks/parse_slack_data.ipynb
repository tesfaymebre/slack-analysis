```python
def get_top_20_user(data, channel='random', user_col='user'):
    """Plot top and bottom message senders for a channel."""
    counts = data[user_col].value_counts()

    counts.head(20).plot.bar(figsize=(15, 7.5))
    plt.title(f'Top 20 Message Senders in #{channel}')
    plt.xlabel('Sender')
    plt.ylabel('Frequency')
    plt.show()

    counts.tail(10).plot.bar(figsize=(15, 7.5))
    plt.title(f'Bottom 10 Message Senders in #{channel}')
    plt.xlabel('Sender')
    plt.ylabel('Frequency')
    plt.show()


def draw_wordcloud(msg_content, week):
    """Create a word cloud from message text."""
    all_words = ' '.join(str(text) for text in msg_content)
    word_cloud = WordCloud(
        background_color='#975429',
        width=500,
        height=300,
        random_state=21,
        max_words=500,
        stopwords=set(stopwords.words('english')),
    ).generate(all_words)

    plt.figure(figsize=(15, 7.5))
    plt.imshow(word_cloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'WordCloud for {week}')
    plt.show()
```
